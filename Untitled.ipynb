```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the diagrams
w = 10  # Wage rate
g = 5000  # Nonlabor income
T = 24  # Total time endowment in hours

# Define leisure and consumption for the budget line without EITC (straight line)
L_no_EITC = np.linspace(0, T, 100)  # Leisure (from 0 to 24 hours)
C_no_EITC = g + w * (T - L_no_EITC)  # Budget line equation without EITC

# Define leisure and consumption for the budget line with EITC (3 segments)
# Phase-in range, plateau range, phase-out range
L_EITC = np.linspace(0, T, 100)

# Phase-in segment: Subsidy increases as income increases
subsidy_rate = 0.4  # 40% subsidy in the phase-in range
C_phase_in = g + (w * (T - L_EITC)) * (1 + subsidy_rate)  # Budget line in phase-in

# Plateau range (flat subsidy rate)
C_plateau = g + w * (T - L_EITC) + 4000  # Assume subsidy reaches $4000

# Phase-out range (subsidy decreases as income increases)
subsidy_decrease = 0.3  # 30% subsidy phase-out
C_phase_out = g + w * (T - L_EITC) + 4000 - subsidy_decrease * (T - L_EITC)

# Plotting
plt.figure(figsize=(10, 6))

# Plot Budget line without EITC
plt.plot(L_no_EITC, C_no_EITC, label="Budget Line Without EITC", color='blue')

# Plot Budget line with EITC (3 segments)
plt.plot(L_EITC, C_phase_in, label="Phase-In Range (EITC)", color='green')
plt.plot(L_EITC, C_plateau, label="Plateau Range (EITC)", color='orange')
plt.plot(L_EITC, C_phase_out, label="Phase-Out Range (EITC)", color='red')

# Adding labels
plt.axvline(x=T, color='black', linestyle='--', label='Max Leisure (24 hours)')
plt.axhline(y=g, color='black', linestyle='--', label='Nonlabor Income (g)')

# Marking Indifference Curves (Tangent Points)
# For simplicity, I'll mark two indifference curves: one near all leisure, one with some labor
indifference_curve1 = g + 5000  # Indifference curve through all leisure
indifference_curve2 = g + 7000  # Indifference curve with labor supplied

plt.plot(L_no_EITC, indifference_curve1, label="Indifference Curve 1 (All Leisure)", linestyle='--', color='purple')
plt.plot(L_EITC, indifference_curve2, label="Indifference Curve 2 (With Labor)", linestyle='--', color='brown')

# Adding titles and labels
plt.title('Labor-Leisure Diagram with and without EITC')
plt.xlabel('Leisure (Hours)')
plt.ylabel('Consumption ($)')
plt.legend()
plt.grid(True)
plt.xlim(0, T)
plt.ylim(0, 12000)

plt.show()
```
